```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Load data with indicators and metrics
df = pd.read_csv('../data/stock_metrics.csv', parse_dates=['Date'])
```

```python
# Function to plot indicators for a single stock
def plot_stock_indicators(ticker, df):
    df_ticker = df[df['Stock'] == ticker]
    
    # Plot 1: Close Price and SMA
    plt.figure(figsize=(12, 6))
    plt.plot(df_ticker['Date'], df_ticker['Close'], label='Close Price', color='blue')
    plt.plot(df_ticker['Date'], df_ticker['SMA_20'], label='20-day SMA', color='orange')
    plt.title(f'{ticker} Close Price and 20-day SMA')
    plt.xlabel('Date')
    plt.ylabel('Price (USD)')
    plt.legend()
    plt.savefig(f'../notebooks/{ticker}_close_sma.png')
    plt.close()

    # Plot 2: RSI
    plt.figure(figsize=(12, 4))
    plt.plot(df_ticker['Date'], df_ticker['RSI_14'], label='14-day RSI', color='purple')
    plt.axhline(70, color='red', linestyle='--', label='Overbought (70)')
    plt.axhline(30, color='green', linestyle='--', label='Oversold (30)')
    plt.title(f'{ticker} 14-day RSI')
    plt.xlabel('Date')
    plt.ylabel('RSI')
    plt.legend()
    plt.savefig(f'../notebooks/{ticker}_rsi.png')
    plt.close()

    # Plot 3: MACD
    plt.figure(figsize=(12, 6))
    plt.plot(df_ticker['Date'], df_ticker['MACD'], label='MACD', color='blue')
    plt.plot(df_ticker['Date'], df_ticker['MACD_Signal'], label='Signal Line', color='orange')
    plt.bar(df_ticker['Date'], df_ticker['MACD_Hist'], label='MACD Histogram', color='gray')
    plt.title(f'{ticker} MACD')
    plt.xlabel('Date')
    plt.ylabel('MACD')
    plt.legend()
    plt.savefig(f'../notebooks/{ticker}_macd.png')
    plt.close()

    # Plot 4: Daily Returns
    plt.figure(figsize=(12, 4))
    plt.plot(df_ticker['Date'], df_ticker['Daily_Return'], label='Daily Return', color='green')
    plt.title(f'{ticker} Daily Returns')
    plt.xlabel('Date')
    plt.ylabel('Return')
    plt.legend()
    plt.savefig(f'../notebooks/{ticker}_returns.png')
    plt.close()
```

```python
# Plot for specified tickers
tickers = ['AAPL', 'AMZN', 'GOOG', 'META', 'MSFT', 'NVDA', 'TSLA']
for ticker in tickers:
    if ticker in df['Stock'].values:
        plot_stock_indicators(ticker, df)
    else:
        print(f"Warning: {ticker} not found in stock_metrics.csv")
print('Visualizations saved in notebooks/ folder')
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load data with indicators and metrics
df = pd.read_csv('../data/stock_metrics.csv', parse_dates=['Date'])

# Function to plot indicators for a single stock
def plot_stock_indicators(ticker, df):
    df_ticker = df[df['Stock'] == ticker]
    
    # Plot 1: Close Price and SMA
    plt.figure(figsize=(12, 6))
    plt.plot(df_ticker['Date'], df_ticker['Close'], label='Close Price', color='blue')
    plt.plot(df_ticker['Date'], df_ticker['SMA_20'], label='20-day SMA', color='orange')
    plt.title(f'{ticker} Close Price and 20-day SMA')
    plt.xlabel('Date')
    plt.ylabel('Price (USD)')
    plt.legend()
    plt.show()  # Display the plot

    # Plot 2: RSI
    plt.figure(figsize=(12, 4))
    plt.plot(df_ticker['Date'], df_ticker['RSI_14'], label='14-day RSI', color='purple')
    plt.axhline(70, color='red', linestyle='--', label='Overbought (70)')
    plt.axhline(30, color='green', linestyle='--', label='Oversold (30)')
    plt.title(f'{ticker} 14-day RSI')
    plt.xlabel('Date')
    plt.ylabel('RSI')
    plt.legend()
    plt.show()  # Display the plot

    # Plot 3: MACD
    plt.figure(figsize=(12, 6))
    plt.plot(df_ticker['Date'], df_ticker['MACD'], label='MACD', color='blue')
    plt.plot(df_ticker['Date'], df_ticker['MACD_Signal'], label='Signal Line', color='orange')
    plt.bar(df_ticker['Date'], df_ticker['MACD_Hist'], label='MACD Histogram', color='gray')
    plt.title(f'{ticker} MACD')
    plt.xlabel('Date')
    plt.ylabel('MACD')
    plt.legend()
    plt.show()  # Display the plot

    # Plot 4: Daily Returns
    plt.figure(figsize=(12, 4))
    plt.plot(df_ticker['Date'], df_ticker['Daily_Return'], label='Daily Return', color='green')
    plt.title(f'{ticker} Daily Returns')
    plt.xlabel('Date')
    plt.ylabel('Return')
    plt.legend()
    plt.show()  # Display the plot

# Plot for specified tickers
tickers = ['AAPL', 'AMZN', 'GOOG', 'META', 'MSFT', 'NVDA', 'TSLA']
for ticker in tickers:
    if ticker in df['Stock'].values:
        plot_stock_indicators(ticker, df)
    else:
        print(f"Warning: {ticker} not found in stock_metrics.csv")
print('Plots displayed for all available tickers')
```
